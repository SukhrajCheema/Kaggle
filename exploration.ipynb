```python
def getData(file_path=DATA_PATH):

    # Returns a df
    return  pd.read_csv(file_path)
```

We immediatley observe a disparity in pay levels, with Male average pay considerably higher than Womens. Lets plot a time series to see this more clearly. 

```python
# Create an empty plot
plt.figure(figsize=(10,6))

# Plot series
plt.plot(genderData['year'], genderData['men_average'], label = "Men's Average", linestyle = '-', marker = 'o')
plt.plot(genderData['year'], genderData['women_average'], label = "Women's Average", linestyle = '-', marker = '^')

# Customize the plot
plt.title("Men's Average Pay vs Women's Average Pay")
plt.xlabel('Year')
plt.ylabel('Values')
plt.grid(True)
plt.legend()

# Ensure all elements fit plot
plt.tight_layout() 
```
